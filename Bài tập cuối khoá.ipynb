```python
from tkinter import *
from tkinter.font import Font
import math
screen=Tk()

screen.geometry('320x265')
screen.title('Máy tính')

fts=Font(family='Tohoma',size=12)

manhinh=Label(screen,bg='#33FFFC',width=320,height=3,font=fts)
manhinh.pack(side=TOP)

fr=Frame(screen)
fr.pack()


bieuthuc=''

def xuly(num):
    global bieuthuc
    bieuthuc=bieuthuc+str(num)
    manhinh.config(text=bieuthuc)

def TinhToan():
    global ketqua
    global bieuthuc
    ketqua=eval(bieuthuc)
    manhinh.config(text=ketqua)
    bieuthuc=''

def Xoa():
    global bieuthuc
    manhinh.config(text='')
    bieuthuc=''
    print(bieuthuc)


def sqrt():
    num = float(bieuthuc)
    result = math.sqrt(num)
    manhinh.config(text=result)
    print(result)

#Hàng 1
bt7=Button(fr,text='7',font=fts,width= 8,pady=5,command=lambda:xuly('7'))
bt7.grid(row=1,column=1)
bt8=Button(fr,text='8',font=fts,width=8,pady=5,command=lambda:xuly('8'))
bt8.grid(row=1,column=2)
bt9=Button(fr,text='9',font=fts,width=8,pady=5,command=lambda:xuly('9'))
bt9.grid(row=1,column=3)
btCong=Button(fr,text='+',font=fts,width=8,pady=5,command=lambda:xuly('+'))
btCong.grid(row=1,column=4)

#Hàng 2
bt4=Button(fr,text='4',font=fts,width=8,pady=5,command=lambda:xuly('4'))
bt4.grid(row=2,column=1)
bt5=Button(fr,text='5',font=fts,width=8,pady=5,command=lambda:xuly('5'))
bt5.grid(row=2,column=2)
bt6=Button(fr,text='6',font=fts,width=8,pady=5,command=lambda:xuly('6'))
bt6.grid(row=2,column=3)
btTru=Button(fr,text='-',font=fts,width=8,pady=5,command=lambda:xuly('-'))
btTru.grid(row=2,column=4)

#Hàng 3
bt1=Button(fr,text='1',font=fts,width=8,pady=5,command=lambda:xuly('1'))
bt1.grid(row=3,column=1)
bt2=Button(fr,text='2',font=fts,width=8,pady=5,command=lambda:xuly('2'))
bt2.grid(row=3,column=2)
bt3=Button(fr,text='3',font=fts,width=8,pady=5,command=lambda:xuly('3'))
bt3.grid(row=3,column=3)
btNhan=Button(fr,text='*',font=fts,width=8,pady=5,command=lambda:xuly('*'))
btNhan.grid(row=3,column=4)

#Hàng 4
btCanBacHai=Button(fr,text='√',font=fts,width=8,pady=5,command=sqrt)
btCanBacHai.grid(row=4,column=1)
bt0=Button(fr,text='0',font=fts,width=8,pady=5,command=lambda:xuly('0'))
bt0.grid(row=4,column=2)
btCham=Button(fr,text='.',font=fts,width=8,pady=5,command=lambda:xuly('.'))
btCham.grid(row=4,column=3)
btChia=Button(fr,text='/',font=fts,width=8,pady=5,command=lambda:xuly('/'))
btChia.grid(row=4,column=4)

#Hàng 5
btCham=Button(fr,text='(',font=fts,width=8,pady=5,command=lambda:xuly('('))
btCham.grid(row=5,column=1)
btCham=Button(fr,text=')',font=fts,width=8,pady=5,command=lambda:xuly(')'))
btCham.grid(row=5,column=2)
btC=Button(fr,text='C',font=fts,width=8,pady=5,command=Xoa)
btC.grid(row=5,column=3)
btBang=Button(fr,text='=',font=fts,width=8,pady=5,command=TinhToan)
btBang.grid(row=5,column=4)

screen.mainloop()
```
